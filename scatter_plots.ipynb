# Plots for each location

```python
import pandas as pd
import numpy as np

df = pd.read_csv('temperature_202501130042.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

## Brian Head, UT

```python
brian_head = df.query("city_name == 'Brian Head'")

# Brian Head Scatter Plot
brian_head.plot(kind = 'scatter', x = 'timestamp', y = 'curr_temperature')

# Standard deviation of temperature for Brian Head
bh_stdev = brian_head['curr_temperature'].std()
print(f"Standard deviation of temperature for Brian Head: {bh_stdev}")
```
